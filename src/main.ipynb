```python
import os
data_path = '../pages/'
```

```python
documentation = {'Utility':[],'Text':[]}
for utility_path in os.listdir(data_path):
    doc_path = os.path.join(data_path,utility_path)
    print('UTILITY::', utility_path)
    documentation['Utility'].append(utility_path[:-3])
    current_text = ''
    with open(doc_path,'r') as f:
        for line in f:
            if line.startswith('>'):
                current_text += ' ' + line[1:].strip()
    documentation['Text'].append(current_text.lower().strip())
```

```python
import pandas as pd
df = pd.DataFrame(documentation)
```

```python
df.to_csv(os.path.join(data_path,'utility_descriptions.csv'))
```
